```python
%matplotlib inline

import sys
sys.path.insert(0, '../../')
import BrasilUSP_iGEM_LabLib as LabLib

import numpy as np
import glob
from IPython.display import display, HTML


ColumnNames = { 
    'Rha amp' : {
    'A' : "-control",
    'B' : "+control",
    'C' : "ZeroC",
    'D' : "C1",
    'E' : "C2",
    'F' : "C3",
    'G' : "C4",
    'H' : "LB",
    },
    'Lac amp' : {
    'A' : "-control",
    'B' : "+control",
    'C' : "ZeroC",
    'D' : "C1",
    'E' : "C2",
    'F' : "C3",
    'G' : "C4",
    'H' : "C5",
    }
}

ColumnNames['Lac clf'] = ColumnNames['Lac amp']

Pidxs = ['Rha amp', 'Lac amp', 'Lac clf']
Pcolors = {'Rha amp' : (1.0,0.5,0.3), 'Lac amp' : (0.3,0.5,1.0), 'Lac clf' : (0.3,1.0,0.5)}
Pcolors_light = {'Rha amp' : (1.0,0.8,0.6), 'Lac amp' : (0.6,0.8,1.0), 'Lac clf' : (0.6,1.0,0.8)}


MeaningColNames = {}
for P in Pidxs:
    MeaningColNames[P] = {v : k for k, v in ColumnNames[P].items()}

def allCols(P, labels, r0 = 1, rf = 9):
    if type(labels) != list: 
        labels = [labels]
    cols = []
    for label in labels:
        row  = MeaningColNames[P][label]
        for j in range(r0,rf+1):
            cols.append(row + '0' + str(j))
    return cols
```

```python
FLlist = {}
tidx = {}
for P in Pidxs: 
    FLlist[P] = []
    tidx[P] = []


for P in Pidxs:
    for file in glob.glob("fluorescencia/t1 p"+P+"/*.txt"):
        tidx[P].append( float( file.split(' ')[5].split('.')[0] ) )
        FLlist[P].append(file)

for P in Pidxs: 
    Idx = np.argsort(tidx[P])
    FLlist[P] = np.array(FLlist[P])[Idx]

oddata = {}
for P in Pidxs:
    ODlist = []
    for j in range(10):
        PP = P.split(' ')[0]
        ODlist.append("OD/od p"+PP+" teste 1/OD "+str(j)+" p"+PP+" t1.txt")
    
    oddata[P] = LabLib.ReadFluor_timefmt_nanomed(ODlist, nr_header=2)


fldata = {}
for P in Pidxs:
    fldata[P] = LabLib.ReadFluor_timefmt_nanomed(FLlist[P])

fldata_rfp = {}
for P in Pidxs:
    fldata_rfp[P] = LabLib.ReadFluor_timefmt_nanomed(FLlist[P], datarow=2)
```

```python
for P in Pidxs:
    print '\n\n'
    display(HTML('<p style="font-size: 25pt;"><b>'+P+'</b></p>'))
    for ColN in ColumnNames[P].values():
        print 
        display(HTML('<p style="font-size: 13pt;"><b>--- ' + ColN + '</b></p>'))
        display(HTML('<p style="font-size: 12pt;">- Green</p>'))
        display( fldata[P][allCols(P,ColN)] )
        display(HTML('<p style="font-size: 12pt;">- Red</p>'))
        display( fldata[P][allCols(P,ColN)] )
        display(HTML('<p style="font-size: 12pt;">- OD</p>'))
        display( oddata[P][allCols(P,ColN)] )
```
